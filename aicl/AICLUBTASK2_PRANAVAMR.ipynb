```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import librosa
import librosa.display
import tensorflow as tf
import kagglehub
import random

from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, f1_score, accuracy_score
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization, GlobalAveragePooling2D, Add, Activation, Multiply, Reshape
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam


# 1. CONFIGURATION
DATA_PATH = kagglehub.dataset_download("uwrfkaggler/ravdess-emotional-speech-audio")

SAMPLE_RATE = 22050
DURATION = 3
SAMPLES_PER_TRACK = SAMPLE_RATE * DURATION

EMOTIONS = {
    '01': 'neutral', '02': 'calm', '03': 'happy', '04': 'sad',
    '05': 'angry', '06': 'fearful', '07': 'disgust', '08': 'surprised'
}




# 2. DATA LOADING & PARSING

def load_data(data_path):
    file_list = []
    for root, dirs, files in os.walk(data_path):
        for file in files:
            if file.endswith('.wav'):
                parts = file.split('.')[0].split('-')
                emotion_code = parts[2]
                actor_code = int(parts[6])
                gender = 'Male' if actor_code % 2 == 1 else 'Female'
                emotion = EMOTIONS[emotion_code]

                file_list.append({
                    'path': os.path.join(root, file),
                    'emotion': emotion,
                    'gender': gender,
                    'actor': actor_code
                })
    return pd.DataFrame(file_list)

print("Loading data...")
df = load_data(DATA_PATH)
print(f"Total files loaded: {len(df)}")




# 3. SPLIT DATA FIRST

le = LabelEncoder()
df['label_encoded'] = le.fit_transform(df['emotion'])

# Split: 80% Train, 10% Val, 10% Test
split_seed = 42
sss1 = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=split_seed)
for train_idx, temp_idx in sss1.split(df, df['label_encoded']):
    train_df = df.iloc[train_idx]
    temp_df = df.iloc[temp_idx]

sss2 = StratifiedShuffleSplit(n_splits=1, test_size=0.5, random_state=split_seed)
for val_idx, test_idx in sss2.split(temp_df, temp_df['label_encoded']):
    val_df = temp_df.iloc[val_idx]
    test_df = temp_df.iloc[test_idx]

print(f"Files -> Train: {len(train_df)}, Val: {len(val_df)}, Test: {len(test_df)}")



# 4. AUDIO AUGMENTATION & PROCESSING


def augment_audio(y, sr):
    if random.random() < 0.5:
        rate = np.random.uniform(0.9, 1.1)
        y = librosa.effects.time_stretch(y, rate=rate)
    if random.random() < 0.5:
        steps = np.random.uniform(-2, 2)
        y = librosa.effects.pitch_shift(y, sr=sr, n_steps=steps)
    if random.random() < 0.5:
        shift_range = int(np.random.uniform(-0.1, 0.1) * len(y))
        y = np.roll(y, shift_range)
    if random.random() < 0.5:
        noise_amp = 0.005 * np.random.uniform() * np.amax(y)
        y = y + noise_amp * np.random.normal(size=y.shape)
    return y

def process_audio_file(file_path, augment=False):
    y, sr = librosa.load(file_path, sr=SAMPLE_RATE)
    if augment:
        y = augment_audio(y, sr)

    if len(y) > SAMPLES_PER_TRACK:
        y = y[:SAMPLES_PER_TRACK]
    else:
        padding = SAMPLES_PER_TRACK - len(y)
        y = np.pad(y, (0, padding), mode='constant')

    mel_spec = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128, fmax=8000)
    mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)
    return mel_spec_db

def create_dataset(dataframe, augment=False):
    X_data = []
    y_data = []
    genders = [] # Track gender for analysis later

    for _, row in dataframe.iterrows():
        path = row['path']
        label = row['label_encoded']
        gender = row['gender']

        # Original
        spec = process_audio_file(path, augment=False)
        X_data.append(spec)
        y_data.append(label)
        genders.append(gender)

        # Augmented (Train only)
        if augment:
            spec_aug = process_audio_file(path, augment=True)
            X_data.append(spec_aug)
            y_data.append(label)
            genders.append(gender)

    return np.array(X_data), np.array(y_data), np.array(genders)

print("\n--- Generating Training Data ---")
X_train, y_train_idx, _ = create_dataset(train_df, augment=True)

print("\n--- Generating Validation Data ---")
X_val, y_val_idx, _ = create_dataset(val_df, augment=False)

print("\n--- Generating Test Data ---")
X_test, y_test_idx, test_genders = create_dataset(test_df, augment=False)

# Normalize
mean = np.mean(X_train)
std = np.std(X_train)
X_train = (X_train - mean) / std
X_val = (X_val - mean) / std
X_test = (X_test - mean) / std

# Reshape
X_train = X_train[..., np.newaxis]
X_val = X_val[..., np.newaxis]
X_test = X_test[..., np.newaxis]

# Categorical Labels
y_train = to_categorical(y_train_idx)
y_val = to_categorical(y_val_idx)
y_test = to_categorical(y_test_idx)





# 5. SE-RESNET MODEL ARCHITECTURE



def squeeze_excite_block(tensor, ratio=16):
    #Squeeze and Excitation Block (Attention Mechanism)

    filters = tensor.shape[-1]
    se = GlobalAveragePooling2D()(tensor)
    se = Reshape((1, 1, filters))(se)
    se = Dense(filters // ratio, activation='relu', kernel_initializer='he_normal', use_bias=False)(se)
    se = Dense(filters, activation='sigmoid', kernel_initializer='he_normal', use_bias=False)(se)
    return Multiply()([tensor, se])

def resnet_block(x, filters, kernel_size=3, stride=1, conv_shortcut=True):

    #A residual block with an SE block added.

    shortcut = x

    if conv_shortcut:
        shortcut = Conv2D(filters, 1, strides=stride)(x)
        shortcut = BatchNormalization()(shortcut)

    # First Conv
    x = Conv2D(filters, kernel_size, strides=stride, padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    # Second Conv
    x = Conv2D(filters, kernel_size, padding='same')(x)
    x = BatchNormalization()(x)

    # Add SE Block (Attention)
    x = squeeze_excite_block(x)

    # Add Shortcut
    x = Add()([x, shortcut])
    x = Activation('relu')(x)
    return x

def build_se_resnet(input_shape, num_classes):
    inputs = Input(shape=input_shape)

    # Initial Conv
    x = Conv2D(64, 7, strides=2, padding='same')(inputs)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling2D(3, strides=2, padding='same')(x)

    # ResNet Blocks
    x = resnet_block(x, 64, stride=1)
    x = resnet_block(x, 64, stride=1, conv_shortcut=False)

    x = resnet_block(x, 128, stride=2)
    x = resnet_block(x, 128, stride=1, conv_shortcut=False)

    x = resnet_block(x, 256, stride=2)
    x = resnet_block(x, 256, stride=1, conv_shortcut=False)

    # Final Classification
    x = GlobalAveragePooling2D()(x)
    x = Dense(256, activation='relu')(x)
    x = Dropout(0.5)(x)
    outputs = Dense(num_classes, activation='softmax')(x)

    model = Model(inputs, outputs, name="SE-ResNet")
    return model

model = build_se_resnet(X_train.shape[1:], len(le.classes_))
model.compile(optimizer=Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()



# 6. TRAINING

save_best = tf.keras.callbacks.ModelCheckpoint("best_se_resnet.keras", monitor='val_accuracy', save_best_only=True, verbose=1)
early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=12, restore_best_weights=True)
reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=1e-6)

history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=50,
    batch_size=32,
    callbacks=[early_stop, reduce_lr, save_best],
    verbose=1
)



# 7. EVALUATION & GENDER BIAS ANALYSIS



print("\n--- Overall Evaluation ---")
y_pred_probs = model.predict(X_test)
y_pred = np.argmax(y_pred_probs, axis=1)
y_true = np.argmax(y_test, axis=1)

print(f"Overall Accuracy: {accuracy_score(y_true, y_pred):.4f}")

# --- Gender Bias Analysis ---
print("\n" + "="*40)
print(" GENDER BIAS ANALYSIS")
print("="*40)

# Create a DataFrame for the test results
results_df = pd.DataFrame({
    'True Label': y_true,
    'Predicted Label': y_pred,
    'Gender': test_genders
})

# Decode labels for better readability
results_df['True Emotion'] = le.inverse_transform(results_df['True Label'])
results_df['Predicted Emotion'] = le.inverse_transform(results_df['Predicted Label'])

# Split by Gender
male_results = results_df[results_df['Gender'] == 'Male']
female_results = results_df[results_df['Gender'] == 'Female']

# Calculate Metrics
male_acc = accuracy_score(male_results['True Label'], male_results['Predicted Label'])
female_acc = accuracy_score(female_results['True Label'], female_results['Predicted Label'])

print(f"\nMale Accuracy:   {male_acc:.4f}")
print(f"Female Accuracy: {female_acc:.4f}")
print(f"Bias Gap:        {abs(male_acc - female_acc):.4f}")



# Plot Side-by-Side Confusion Matrices
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Male CM
cm_male = confusion_matrix(male_results['True Label'], male_results['Predicted Label'])
sns.heatmap(cm_male, annot=True, fmt='d', cmap='Blues', ax=axes[0],
            xticklabels=le.classes_, yticklabels=le.classes_)
axes[0].set_title(f"Male Confusion Matrix (Acc: {male_acc:.2f})")
axes[0].set_xlabel('Predicted')
axes[0].set_ylabel('True')

# Female CM
cm_female = confusion_matrix(female_results['True Label'], female_results['Predicted Label'])
sns.heatmap(cm_female, annot=True, fmt='d', cmap='Reds', ax=axes[1],
            xticklabels=le.classes_, yticklabels=le.classes_)
axes[1].set_title(f"Female Confusion Matrix (Acc: {female_acc:.2f})")
axes[1].set_xlabel('Predicted')
axes[1].set_ylabel('True')

plt.tight_layout()
plt.show()

# Classification Reports
print("\n--- MALE Classification Report ---")
print(classification_report(male_results['True Label'], male_results['Predicted Label'], target_names=le.classes_))

print("\n--- FEMALE Classification Report ---")
print(classification_report(female_results['True Label'], female_results['Predicted Label'], target_names=le.classes_))
```
